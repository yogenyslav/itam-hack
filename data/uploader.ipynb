```python
import pandas as pd
import httpx
import random
```

```python
server_url = 'http://localhost:9999'
client = httpx.Client()
names = pd.read_csv('russian_names.csv', sep=';')
surnames = pd.read_csv('russian_surnames.csv', sep=';')
```

```python
async def create_user(index: int):
    """
    {
        "email": "test99@test.com",
        "password": "test123456",
        "first_name": "Денис",
        "last_name": "Трипольский",
        "internal_role": "admin"
    }
    """
    # select random name and surname
    random_name_index = random.randint(0, len(names['Name']) - 1)
    random_surname_index = random.randint(0, len(surnames['Surname']) - 1)
    name = names['Name'][random_name_index]
    surname = surnames['Surname'][random_surname_index]
    print(name, surname)
    response = client.post(
        server_url +"/auth/signup"
        , json={
            "internal_role": "student",
            "email": f"test{index}@test.com",
            "password": "test123456",
            "first_name": name,
            "last_name": surname,
        }
    )
    print(response.json())
         
            
            
```

```python
# create 100 users with index from 100 to 200 in async mode all together

tasks = [create_user(i) for i in range(100, 200)]

import asyncio

asyncio.gather(*tasks)
```
